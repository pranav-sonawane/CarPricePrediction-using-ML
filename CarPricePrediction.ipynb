```python
corollas_df = pd.read_csv('ToyotaCorolla.csv')
corollas_df.head(5)
```

```python
corollas_df= corollas_df.dropna(axis=0,how='all')
print(corollas_df.shape)
```

```python
corollas_df.head(10)
```

```python
corollas_df.isnull().sum()
```

```python
corollas_df=corollas_df.drop(["Mistlamps"], axis = 1)
```

```python
print(corollas_df['Power_Steering'].value_counts())
print(corollas_df['Radio'].value_counts())
print(corollas_df['Sport_Model'].value_counts())
print(corollas_df['Metallic_Rim'].value_counts())
corollas_df['Powered_Windows'].value_counts()
```

```python
corollas_df= corollas_df.fillna(value={"Color":corollas_df['Color'].mode()[0],"CC":corollas_df['CC'].mean()})
corollas_df.isnull().sum()

```

```python
corollas_df= corollas_df.dropna(axis=0,how='any')
```

```python
corollas_df.isnull().sum()
```

```python
predictors_df_corollas = corollas_df[['Age_08_22', 'KM', 'Fuel_Type','Color', 'HP', 'Automatic', 'CC',
              'Doors', 'Mfr_Guarantee', 'Airco', 'Sport_Model', 'ABS', 'Airbag_1', 'Airbag_2', 'Powered_Windows','Power_Steering','Radio','Sport_Model','Metallic_Rim']]
response_corollas = corollas_df['Price']
```

```python
predictors_df_corollas = pd.get_dummies(predictors_df_corollas)
predictors_df_corollas.corr()
```

```python
predictors_df_corollas = predictors_df_corollas.drop(["Fuel_Type_Diesel"], axis = 1)
predictors_df_corollas.corr()
```

```python
z_score_norm = preprocessing.StandardScaler()
predictor_df_normalized = z_score_norm.fit_transform(predictors_df_corollas)
predictor_df_normalized = pd.DataFrame(predictor_df_normalized, columns = predictors_df_corollas.columns)
predictor_df_normalized.head(10)
```

## Multiple Linear Regression Model

```python
X = predictor_df_normalized
y = response_corollas
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
linear_model = LinearRegression()
linear_model = linear_model.fit(train_X, train_y)
print('intercept ', linear_model.intercept_)
print(pd.DataFrame({'Predictor': X.columns, 'coefficient': linear_model.coef_}))
```

```python
predicted_y_training = linear_model.predict(train_X)

regressionSummary(train_y, predicted_y_training)
```

```python
print(adjusted_r2_score(train_y, predicted_y_training, linear_model))
```

```python
predicted_y_test = linear_model.predict(test_X)  # these are our y_hat values!!

result = pd.DataFrame({'Predicted': predicted_y_test, 'Actual': test_y,
                       'Residual': test_y - predicted_y_test})
result.head()
```

```python
regressionSummary(test_y, predicted_y_test)
```

## Random Forest Model

```python
RF_MNIST = RandomForestRegressor(random_state=11,max_depth=7,n_estimators=150,min_samples_split=2).fit(train_X, train_y)
predicted_y_training_MNIST = RF_MNIST.predict(train_X)

print(predicted_y_training_MNIST)
regressionSummary(train_y,predicted_y_training_MNIST)
```

```python
predicted_y_test = RF_MNIST.predict(test_X)

regressionSummary(test_y, predicted_y_test)
```

## Neural Network Model

```python
X = predictor_df_normalized
y = response_corollas
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state=13)
```

```python
NN = MLPClassifier(hidden_layer_sizes=(100,100), activation="logistic", solver="adam",
                   learning_rate_init=0.01, verbose=True, max_iter=100, batch_size=10).fit(train_X, train_y)
predicted_y_training = NN.predict(train_X)

print(classification_report(train_y, predicted_y_training))
```

```python
regressionSummary(train_y, predicted_y_training)
```

```python
predicted_y_test = NN.predict(test_X)

regressionSummary(test_y, predicted_y_test)
```
